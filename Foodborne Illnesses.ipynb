```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in the data file
dff = pd.read_csv("outbreaks.csv")
```

```python
# find illnesses count by each month
group_month = dff.groupby('Month',sort=False).Illnesses.count()
ax = group_month.plot(kind='bar', color='g')
rects = ax.patches
labels = [group_month.iloc[i] for i in range(len(rects))]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 40, label, ha='center', va='bottom',rotation='vertical')
plt.ylabel('Count of Illnesses')
plt.title('Foodborne Illnesses By Month')
plt.yticks([400,800,1200,1600,2000,2400])
plt.show()
```

```python
#find the trend in illnesses count by year
group_year = dff.groupby('Year',sort=False).Illnesses.count()
ax = group_year.plot(kind='bar', color='g')
rects = ax.patches
labels = [group_year.iloc[i] for i in range(len(rects))]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 40, label, ha='center', va='bottom',rotation='vertical')
plt.ylabel('Count of Illnesses')
plt.title('Foodborne Illnesses By Year')    
plt.yticks([400,800,1200,1600,2000,2400])
plt.show()
```

```python
#find the top ten states 
group_state = dff.groupby('State',sort=False).Illnesses.count()
group_state.sort_values(ascending=False,inplace=True)
top10 = group_state.head(10)

ax = top10.plot(kind='bar', color='g')
rects = ax.patches
labels = [top10.iloc[i] for i in range(len(rects))]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')
plt.ylabel('Count of Illnesses')
plt.title('Top Ten States With Foodborne Illnesses During 1998-2015')    
plt.show()
```

```python
#find the top 10 locations
group_location = dff.groupby('Location',sort=False).Illnesses.count()
group_location.sort_values(ascending=False,inplace=True)
top10 = group_location.head(10)
ax = top10.plot(kind='bar', color='g')
rects = ax.patches
labels = [top10.iloc[i] for i in range(len(rects))]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')
plt.ylabel('Count of Illnesses')
plt.title('Top Ten Locations of Foodborne Illnesses During 1998-2015')    
plt.show()
```

```python
#find top 10 foods that cause most illnesses
group_food = dff.groupby('Food',sort=False).Illnesses.count()
group_food.sort_values(ascending=False,inplace=True)
top10 = group_food.head(10)
ax = top10.plot(kind='bar', color='g')
rects = ax.patches
labels = [top10.iloc[i] for i in range(len(rects))]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, label, ha='center', va='bottom')
plt.ylabel('Count of Illnesses')
plt.title('Top Ten Types of Food That Caused Foodborne Illnesses During 1998-2015')    
plt.show()
```

```python
#find the top 10 species that cause illnesses
group_species = dff.groupby('Species',sort=False).Illnesses.count()
group_species.sort_values(ascending=False,inplace=True)
top10 = group_species.head(10)

ax = top10.plot(kind='bar', color='g')
rects = ax.patches
labels = [top10.iloc[i] for i in range(len(rects))]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, label, ha='center', va='bottom')
plt.ylabel('Count of Illnesses')
plt.title('Top Ten Species That Caused Foodborne Illnesses During 1998-2015')    
plt.show()
```

```python
# draw the choloropleth map
import plotly.plotly as py
import numpy as np

for col in dff.columns:
    dff[col] = dff[col].astype(str)

group_state = dff.groupby('State',sort=False).Illnesses.count().reset_index()

dff['text'] = dff['State'] + '<br>' +\
    'Illness '+dff['Illnesses']
    
scl = [[0.0, 'rgb(149,249,67)'],[0.2, 'rgb(204,249,67)'],[0.4, 'rgb(237,249,67)'],\
            [0.6, 'rgb(249,227,67)'],[0.8, 'rgb(249,270,67)'],[1.0, 'rgb(249,116,67)']]
s = group_state['State']
data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = s,
        z = group_state['Illnesses'],
        locationmode = 'USA-states',
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Illnesses Count")
        ) ]

layout = dict(
        title = 'Foodborne Illnesses 1998-2015<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

```python
dff['Status'] = dff['Status'].apply(lambda x: 2 if x=='Confirmed' else 1)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import accuracy_score


cols_to_dict = [ 'Year', 'Month', 'State', 'Location', 'Food','Species' ]
category_dict = dff[ cols_to_dict ].to_dict( orient = 'records' )


dff.drop('Ingredient',axis=1,inplace=True)
dff.drop('Serotype/Genotype',axis=1,inplace=True)

dff['Species'].fillna('Norovirus genogroup I',inplace=True)
dff['Location'].fillna('Restaurant',inplace=True)
dff['Food'].fillna('Multiple Foods',inplace=True)
dff['Hospitalizations'].fillna('1',inplace=True)
dff['Fatalities'].fillna('1',inplace=True)

#features_train,features_test,labels_train,labels_test = train_test_split(dff[features],dff[labels], test_size=0.3, random_state=42)
v = DictVectorizer(sparse=False)
X = v.fit_transform(category_dict)
nb_clf = GaussianNB()
features_train,features_test,labels_train,labels_test = train_test_split(X,dff['Status'], test_size=0.3, random_state=42)

nb_clf.fit(features_train,labels_train)
p = nb_clf.predict(features_test)
print 'Accuracy = ',accuracy_score(p,labels_test)
```
